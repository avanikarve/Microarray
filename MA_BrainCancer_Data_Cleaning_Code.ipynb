```python
import pandas as pd
```

```python
df2 = pd.read_csv("D:\\MSc Statistics SPPU\\SEM II\\Microarray\\Assignments\\Brain_5C.csv", index_col=0)
print("Shape: ", df2.shape)
df2 = df2.T
df2.head()
```

```python
first_row = df2.iloc[0]
col_names = df2.columns
```

```python
# ID Type association
ID_type_dict2 = {}
for category in first_row.unique():
    ID_type_dict2[category] = list(col_names[first_row == category])
```

```python
for cancer in first_row.unique():
    print(cancer, len(ID_type_dict2[cancer]))
```

```python
# Reseting column names
df2.columns = df2.iloc[0]
df2 = df2[1:]
unique_cols = list(df2.columns.unique())  

rename_dict2 = {}  
for i, col in enumerate(unique_cols):
    if col == 'normal':
        pass
    else:
        rename_dict2[col] = f"T{i+1}" 

df2 = df2.rename(columns=rename_dict2)  
df2.head()
```

```python
df2.to_csv("C:\\Users\\Avani\\Downloads\\Clean_Brain_5C.csv", index = True)
```

```python
# Saving Association Dictionaries as csv files
pd.DataFrame(list(ID_type_dict2.items()), columns=["Category", "Columns"]).to_csv("C:\\Users\\Avani\\Downloads\\ID_type_dict2.csv", index=False)  
pd.DataFrame(list(rename_dict2.items()), columns=["Original Name", "New Name"]).to_csv("C:\\Users\\Avani\\Downloads\\rename_dict2.csv", index=False)  
```
